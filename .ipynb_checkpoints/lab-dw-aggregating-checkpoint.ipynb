This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by first performing data cleaning, formatting, and structuring.

1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"

df = pd.read_csv(url)
```

```python
df.columns
```

```python
df['Response'].value_counts()
```

```python
new_data = df[(df['Total Claim Amount'] > 1000) & (df['Response'] == 'Yes')]
new_data
```

```python
# 2/ Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers 
# who have responded "Yes" to the last marketing campaign. Write your conclusions.

pivot_df = df.pivot_table(index='Policy Type', columns = 'Gender', values = ['Total Claim Amount'], aggfunc='mean')
pivot_df.round(2)
```

```python
# Jav: the  Male claim more than the Female for Corporate Auto (by a lot) and Personal Auto. For special auto, it is the 
#      female that claim the most
```

```python
# 3/Analyze the total number of customers who have policies in each state, and then filter the results to only include 
# states where there are more than 500 customers.
df.groupby(['State'])['Customer'].count()
```

```python
# Jav: They are all above 500 customers... not done the filter...
```

```python
# 4/Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

pivot_df = df.pivot_table(index='Education', columns = 'Gender', values = ['Customer Lifetime Value'], aggfunc=['max', 'min','median'])
pivot_df.round(2)
```

```python
# Jav: highest Max is for male/high school,the values of max are spread from 32K to 83K
# Median: highest is for male/high school (6.3K), the values of median are relatively close from 5.3K (female/doctor) to 6.3K
# lowest Min is for male/bachelor, the values of min are very close 1.9K to 2.4 K
```
